```python
from nltk.corpus import webtext
```

```python
webtext.fileids()
```

```python
for file in webtext.fileids():
        words = [word.lower()
                 for word in webtext.words(file)]
```

```python
print(file)
```

```python
print(words)
```

```python
for file in webtext.fileids():
        words = [word.upper()
                 for word in webtext.words(file)]
```

```python
print(words)
```
