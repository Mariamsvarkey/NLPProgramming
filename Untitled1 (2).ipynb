```python
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
doc = nlp("Alibaba's Ant Financial-backed online food delivery and restaurant discovery platform Zomato has acquired Uber Eats, the food delivery business of ride-hailing giant Uber India for around Rs 2,485 crore ($350 million) in an all-stock deal. The deal gives the ride-hailing app a 10 per cent stake in Zomato. Notably, Zomato will not absorb Uber Eat employees, which means they will either be absorbed in Uber's other verticals or could face lay-off. There's little clarity on their future roles right now.")
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
for token in doc:
    print(token.text)
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
for token in doc:
    print(token.text, token.has_vector, token.vector_norm, token.is_oov)
```
